## Grid printing classes (simple inheritance, overriding functions)

```python
class Grid:
    def __init__(self, name:str):
        self._name = name
    
    def print(self):
        print(f"\n{self._name}:")


class XGrid(Grid):
    def __init__(self, name:str, size:int):
        super().__init__(name)
        self._size = size

    def print(self):
        super().print()
        for r in range(self._size):
            line = ""
            for c in range(self._size):
                if r==c or r+c == self._size-1:
                    line += "#"
                else:
                    line += "."
            print(line)

class RectGrid(Grid):
    def __init__(self, name:str, colSize:int, rowSize:int):
        super().__init__(name)
        self._c_olSize = self.colSize
        self.rowSize = self.rowSize

    def print(self):
        super().print()
        for r in range(self.rowSize):
            print("*" * self.colSize)
```

```python
grids = [ XGrid(name="Small X", size=3), Grid(name="Nothing"), XGrid(name="Large X", size=9) ]
for g in grids:
    g.print()
```

## Matrix class (encapsulation, incremental code development, lambdas, raising exceptions)

```python
class Matrix: 
    def __init__(self, dims: tuple[int,int], val: int = 0):
        self._dims = dims
        self._data = [val] * (dims[0] * dims[1])
        self._val = val

    def _pos2idx(self, pos: tuple[int,int]) -> int:
        r, c = pos
        rows, cols = self._dims
        if not (0 <= r < rows and 0 <= c < cols):
            raise IndexError(f"position {pos} out of bounds for dims {self._dims}")
        return r * cols + c

    def set(self, pos: tuple[int, int], val: int) -> None:
        """set matrix element at row/col to value"""
        idx = self._pos2idx(pos)
        self._data[idx] = val

    def get(self, pos: tuple[int, int]) -> int:
        """return matrix element at row/col"""
        idx = self._pos2idx(pos)
        return self._data[idx]
    
    def print(self):
        print(set)
```

Phase 1

```python
# ----- phase 1 -----
m = Matrix(dims=(2,3))
m._data
m._dims
```

```python
# ----- phase 2 -----
m = Matrix(dims=(2,3))
m.set( pos=(0,0), val=10 )
m.set( pos=(1,2), val=20 )


m.get( pos=(1,2) )
m.get( pos=(0,1) )  
```
