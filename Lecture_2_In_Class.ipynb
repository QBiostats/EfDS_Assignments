## Dictionaries

```python
licensePlates = ["AN", "BO", "CH", "DA"]
colors=["white", "black", "white", "white"]

# licencePlate2Color
m =dict(zip(licensePlates, colors))
```

```python
m["AN"] # get the color of the license plate "AN"
```

```python
len(m) # number of key-value pairs in the dictionary
```

```python
m
```

```python
m["XX"]
```

```python
m.get("AN", "unknown")
```

```python
list(m.values())
```

```python
list(m.keys())
```
