## Generating random integers

```python
from random import randint
```

```python
randint(0,9)
```

```python
nums= tuple(range(20))
vs = [randint(0,9) for n in nums]
vs
```

```python
set(vs)
```

## Counting elements

```python
v2cnt = {}

unique_vs = set(vs)

for n in unique_vs:
    v2cnt[n] = 0

for n in vs:
    v2cnt[n] += 1
```

```python
v2cnt
```

## Build a dictionary with values being lists

```python
from random import sample

role2num = { "setter": 1, "dia": 1, "middle": 2, "outside": 2 }

player2role = { 
    "Chen": "setter", "Martijn": "setter",
    "Marnick": "dia", "Simon": "dia",
    "David": "middle", "Luuk": "middle",
    "Ronald": "outside", "Alex": "outside", "Kadir": "outside", "Koen": "outside"
}
```

```python
player2role
```

```python
role2players = {}
for player, role in player2role.items():
    if role not in role2players:
        role2players[role] = []
    role2players[role].append(player)
```

```python
role2players
```

```python
chosen_team= {}

for role, num in role2num.items():
    available_players = role2players[role]
    chosen_team[role] = sample(available_players, num)
```

```python
chosen_team
```

## Read weather data

```python
import urllib.request, json 
url = "https://api.open-meteo.com/v1/forecast?latitude=52.16&longitude=4.49&hourly=temperature_2m"
data = json.load( urllib.request.urlopen(url) )
# print(data)          # use this to print the nested list with data
```

```python
print(data)
```

```python
temps = data["hourly"]["temperature_2m"]
print(temps)
```

```python
times = data["hourly"]["time"]
print(times)
```

```python
time2temp = dict(zip(times,temps))
print(time2temp)
```
