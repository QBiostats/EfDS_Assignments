## A function to convert an exam score to a Dutch grade

```python
def score2grade(score, maxScore):
    """
    Linear transformation of a score to a grade
    """
    grade = 1 + (score / maxScore) * 9
    print(f"final grade: {grade}")
    if score == 0:
        return 1
    elif score == maxScore:
        return 10
    elif score > maxScore:
        raise ValueError("score must be between 0 and maxScore")
```

```python
score2grade(4,10)
```

## A lambda function to convert score to grade

```python
s2g = lambda score,maxScore:  1 + (score / maxScore) * 9
```

```python
s2g(4,10)
```

## A condition in a loop in a loop: painting with dots

```python
def printGrid(n):
    for r in range(n):
        print(n * ".")
    
```

```python
printGrid(2)
```

```python
def printBacksLash(n):
    for r in range (n):
        for c in range(n):
            if r == c:
                print("#",end="")
            else:
                print(".",end="")
        print()
```

```python
printBacksLash(6)
```

```python
def printSlash(n):
    for r in range(n):
        for c in range(n):
            if r + c == n - 1:
                print("#", end="")
            else:
                print(".", end="")
        print()
```

```python
printSlash(8)
```

```python
def printSquare(n):
    for r in range(n):
        for c in range(n):
            if r == 0 or r == n-1 or c==0 or c== n-1:
                print("#",end="")
            else:
                print(".", end="")
        print()
```

```python
printSquare(7)
```

```python
def printX(n):
    for r in range(n):
        for c in range(n):
            if r == c or r + c == n -1:
                print("#", end="")
            else:
                print(".", end="")
        print()
```

```python
printX(10)
```

```python
def printLUTriangle(n):
    for r in range(n):
        for c in range(n):
            if r == 0 or c == 0 or r + c == n-1:
                print("#", end="")
            elif r + c < n-1:
                print(".", end="")
        print()
```

```python
printLUTriangle(10)
```

```python
def printRBTriangle(n):
    for r in range(n):
        for c in range(n):
            if r == n-1 or r+c == n-1 or c == n-1:
                print("#",end="")
            elif r+c < n-1:
                print(" ", end="")
            else:
                print(".",end="")
        print()
```

```python
printRBTriangle(10)
```

## Processing lines of a romote file

```python
import urllib.request
```

```python
url = "https://raw.githubusercontent.com/LUMC/EfDS/main/README.md"

lines = urllib.request.urlopen(url).readlines()
```

```python
textsWithCourse = []

for l in lines:
    # 1. Zet de 'bytes' om naar een normale tekst-string (str)
    t = l.decode("utf-8")
    
    # 2. Check of het woord "course" in de tekst staat
    if "course" in t:
        # 3. Voeg de regel toe aan onze lijst
        textsWithCourse.append(t)

# Print het resultaat om te controleren
print(textsWithCourse)
```
