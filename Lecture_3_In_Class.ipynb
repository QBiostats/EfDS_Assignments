## Flow control

if section

```python
response = input(prompt="Give me your age") # input function, prompt is the string that is shown to the user. The input function always returns a string.
age = int(response) # convert the string to an integer. If the user gives a non-integer, this will raise an error.
print(age)
print(type(age)) # the type of the response is string, even if the user gives a number.
```

```python
response = input(prompt="Give me your age")
age = int(response)
if age >= 0 and age < 125:
    print("------")
    print("Age is correct!")
    print("------") # only does this block of code if age is the range 0 - 125, else it will skip over it and only print "next line"

print("Next line")
```

```python
response = input(prompt="Give me your age")
age = int(response)
if age >= 0 and age < 125:
    print("------")
    print("Age is correct!")
    print("------") 
else:
    print("------")
    print(f"Age '{age}' is NOT CORRECT!")
    print("------") # if age is not in the range 0 - 125, it will print this block of code.

print("Next line")
```

```python
response = input(prompt="Give me your age")
age = int(response)
if age > 0 and age < 125:
    print("------")
    print("Age is correct!")
    print("------") 
elif age == 0:
    print("Age is exactly zero")
else:
    print("------")
    print(f"Age '{age}' is NOT CORRECT!")
    print("------") 

print("Next line")
```

## loops

```python
names = ["Claire","Phil", "Haley", "Cameron", "Luke", "Michell", "Jay", "Gloria"]

for n in names: 
    print(f"Hello, {n}!")
```

```python
for n in range(5):
    print(f"Begin of the loop body for n={n}")
    print(n)
    print(f"End of the loop body for n={n}")
```

```python
for n in range(5):
    print(f"Begin of the loop body for n={n}")
    if n == 1:
        continue # go immediately to the next itteration of the loop, skip the rest of the code in the loop body for n=1.
    print(f"End of the loop body for n={n}")
```

```python
for n in range(5):
    print(f"Begin of the loop body for n={n}")
    if n == 1:
        continue
    if n == 3:
        break # stop the current loop, go outside.
    print(f"End of the loop body for n={n}")
```

```python
for n in range(5):
    print(f"Begin of the loop body for n={n}")
    if n == 1:
        continue
    if n == 3:
        break 
    print(f"End of the loop body for n={n}")
else:
    print("After the loop") # execute if loop finished without "break"
```

```python
from math import log, exp

def myMean(x, isGeom = False):
    if len(x) == 0:
        raise RuntimeError("This function requires a nonempty list!")
    if isGeom:
         log0fX = [log(n) for n in x] # calc log of all nums
         meanLogOfX = sum(log0fX) / len(log0fX) # calc the mean of logs
         return exp(meanLogOfX) # calc exp
    else:
        return sum(x) / len(x)
```

```python
nums = []
myMean(nums)
```

```python
myMean([1,2,4], isGeom=False)
```

```python
myMean([1,2,4], isGeom=True)
```
