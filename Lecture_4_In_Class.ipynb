## Recall

```python
def rowOfPascalsTriangle(n: int) -> list[int]: # n should be of type int
    """ Returns the n-th row of Pascal's triangle"""
    if n < 0:
        raise ValueError("n should be 0 or larger!")

    vs: list[int] = []

    for r in range(0,n+1):
        for c in range(1,r):
            vs[c-1] += vs[c]
        vs.insert(0,1)

    return vs # returns a list
```

```python
rowOfPascalsTriangle(3)
```

```python
rowOfPascalsTriangle(10)
```

## Object oriented programming

### The definition of the class

code written by the coder

```python
class SimpleLamp:
    def __init__(self, location): # this is the constructor of the type SimpleLamp
        self._amITurnedOn = False
        self._location = location

    def turnOn(self):
        self._amITurnedOn = True
        print(f"The lamp {self._location} has been enabled")

    def turnOff(self):
        self._amITurnedOn = False
        print("I have been disabled")

    def isOn(self):
        return self._amITurnedOn
    
    def __str__(self):
        return f"I am the lamp location at '{self._location}' and my state is {self._amITurnedOn}."
```

### The perspective of the user

code written by the user. As a user you only need to know that there is such a function.

```python
# SimpleLamp will be the name of the class
# lamp will be a variable holding an object (instance) of class SimpleLamp

lamp = SimpleLamp(location = "Kitchen")
lamp2 = SimpleLamp(location = "Living room")

# I want to call the method "turnOn" for the "lamp" instance. 
lamp.turnOn()

lamp2.turnOff()

if lamp.isOn():
    print("The lamp in the kitchen is on")
```

```python
l = SimpleLamp("Garden")
print(l)
l.turnOn()
print(l)
l.turnOff()
print(l)
```
