```python
def rowOfPascalsTriangle(n: int) -> list[int]: # n should be of type int
    """ Returns the n-th row of Pascal's triangle"""
    if n < 0:
        raise ValueError("n should be 0 or larger!")

    vs: list[int] = []

    for r in range(0,n+1):
        for c in range(1,r):
            vs[c-1] += vs[c]
        vs.insert(0,1)

    return vs # returns a list
```

```python
rowOfPascalsTriangle(3)
```
